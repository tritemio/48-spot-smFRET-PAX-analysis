```python
save_figures = True
savefigdir_ = 'figures'

def save_name(name, folder='.', meas_label=None, nospaces=True):
    if meas_label is None:
        meas_label = mlabel  # global
    if len(meas_label) > 0 and not meas_label.endswith('_'):
        meas_label = meas_label + '_'
    sname = '%s/%s%s' % (folder, meas_label, name)
    if nospaces:
        sname = sname.replace(' ', '_')
    return sname 
    
def savefig(name, nospaces=True, meas_label=None, **kwargs):
    if not save_figures:
        return
    kwargs_ = dict(dpi=300, bbox_inches='tight')
                   #frameon=True, facecolor='white', transparent=False)
    kwargs_.update(kwargs)
    fname = save_name(name, savefigdir_, meas_label=meas_label, nospaces=nospaces)
    plt.savefig(fname, **kwargs_)
    print('Saved: %s.png' % fname, flush=True)
```

```python
dcr_fname = 'DCR_N2Top_N5Bott.csv'
dcr = pd.read_csv(dcr_fname, index_col=0)
dcr.head()
```

```python
heatmap48(dcr.donor, title='D-SPAD DCR', vmin=0, annot=True)
savefig('DCR donor', meas_label='')
```
